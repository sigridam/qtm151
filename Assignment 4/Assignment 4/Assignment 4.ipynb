- Suppose that 5 students get the following grades:

```python
#(a) Hint: Use a double for loop structure:
#---- for studentgrade in list_studentgrades:
#----     index = 0
#----     for letter in list_letters:
#----          Body
#----          index = index + 1

# (b) Inside the body, use an if/else statement to print the correct letter grade.
# You can make your code more concise by doing similar steps to question (d).

# Write your answer here:

list_studentgrades = [80, 92, 60, 42, 97]

for student_grade in list_studentgrades:
    index = 0
    for letter in list_letters:
        if student_grade>93:
            letter="A"
        elif 87<=student_grade<=92:
            letter="A-"
        elif 83<=student_grade<=86:
            letter="B+"
        elif 80<=student_grade<=82:
            letter="B"
        elif 75<=student_grade<=79:
            letter="B-"
        elif 70<=student_grade<=74:
            letter="C+"
        elif 65<=student_grade<=69:
            letter="C"
        elif 60<=student_grade<=64:
            letter="C-"
        elif 55<=student_grade<=59:
            letter="D"
        else:
            letter= "F"
    print("the student got a numeric grade of "+str(student_grade)+" and their letter grade was "+str(letter))
    index=index+1



```
